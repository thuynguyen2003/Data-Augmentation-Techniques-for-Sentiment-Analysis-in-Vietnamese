```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
data_path = './data/keyword_replace.csv'
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv(data_path)

# Chia data frame thành các phân khúc theo nhãn
negative_samples = df[df['label'] == 'negative']
neutral_samples = df[df['label'] == 'neutral']
positive_samples = df[df['label'] == 'positive']

# Lấy ngẫu nhiên các mẫu từ mỗi phân khúc theo tỷ lệ mong muốn
num_samples_negative = 350
num_samples_neutral = 300
num_samples_positive = 350

selected_negative_samples = negative_samples.sample(n=num_samples_negative, random_state=42)
selected_neutral_samples = neutral_samples.sample(n=num_samples_neutral, random_state=42)
selected_positive_samples = positive_samples.sample(n=num_samples_positive, random_state=42)

# Kết hợp các mẫu đã chọn thành một data frame kết quả
data = pd.concat([selected_negative_samples, selected_neutral_samples, selected_positive_samples]).sample(frac=1)

# Lưu ý rằng 'random_state' được sử dụng để đảm bảo việc lấy ngẫu nhiên có thể tái sản xuất được.

# Hiển thị thông tin về tỷ lệ các nhãn trong mẫu đã chọn
print(data['label'].value_counts())
print("Kích thước của mẫu dữ liệu:", len(data))
data.head()
```

```python
label_dict = {
    'negative':0,
    'neutral' : 1,
    'positive' :2
}
```

```python
X = data[['Tiếng Việt', 'Code-mixed']]
Y = data['label'].map(label_dict)
X_train, X_remain, Y_train, Y_remain = train_test_split(X, Y, test_size= 300, random_state=42)
X_dev, X_test, Y_dev, Y_test = train_test_split(X_remain, Y_remain, test_size= 200, random_state=42)
Xvn_train = X_train['Tiếng Việt']
Xmixed_train = X_train['Code-mixed']

Xvn_test = X_test['Tiếng Việt']
Xmixed_test = X_test['Code-mixed']

Xvn_dev = X_dev['Tiếng Việt']
Xmixed_dev = X_dev['Code-mixed']
```

```python
from datasets import DatasetDict, Dataset
import pickle
# tạo cho bộ Tiếng Viết

# Convert DataFrame to dictionary format
train_dict = DatasetDict({'text': list(Xvn_train), 'label':list(Y_train)})
dev_dict = DatasetDict({'text': list(Xvn_dev), 'label':list(Y_dev)})
test_dict = DatasetDict({'text':list(Xvn_test), 'label':list(Y_test)})

# Create Dataset from the dictionaries
train_dataset = Dataset.from_dict(train_dict)
dev_dataset = Dataset.from_dict(dev_dict)
test_dataset = Dataset.from_dict(test_dict)
# tạo dataset
VietNam = DatasetDict({'train': train_dataset, 'dev':dev_dataset, 'test': test_dataset})

# Lưu file pickle
with open('./data/Final_VietNam.pkl', 'wb') as f:
    pickle.dump(VietNam, f)
```

```python
from datasets import DatasetDict, Dataset
import pickle
# tạo cho bộ Mixed

# Convert DataFrame to dictionary format
train_dict = DatasetDict({'text': list(Xmixed_train), 'label':list(Y_train)})
dev_dict = DatasetDict({'text': list(Xmixed_dev), 'label':list(Y_dev)})
test_dict = DatasetDict({'text':list(Xmixed_test), 'label':list(Y_test)})

# Create Dataset from the dictionaries
train_dataset = Dataset.from_dict(train_dict)
dev_dataset = Dataset.from_dict(dev_dict)
test_dataset = Dataset.from_dict(test_dict)

# tạo dataset
Mixed = DatasetDict({'train': train_dataset, 'dev': dev_dataset, 'test': test_dataset})

# Lưu file pickle
with open('./data/Final_Mixed.pkl', 'wb') as f:
    pickle.dump(Mixed, f)
```
